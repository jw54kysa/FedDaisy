```python
import pandas as pd
import pickle
import numpy as np

testLosses = []
with open("testLosses.pck", "rb") as f:
    x = pickle.load(f)

for i in range(0,len(x)):
    testLosses.append(np.mean(x[i]))

trainLosses = []
with open("trainLosses.pck", "rb") as f:
    x = pickle.load(f)

for i in range(0,len(x)):
    trainLosses.append(np.mean(x[i]))
```

```python
import matplotlib.pyplot as plt

with open("testLosses.pck", "rb") as f:
    testLosses = pickle.load(f)
    
with open("trainLosses.pck", "rb") as f:
    trainLosses = pickle.load(f)

# Generate epochs based on the length of the losses
x_val = range(1, len(testLosses[0]) + 1)

# Plot the loss curves
plt.figure(figsize=(10, 6))
for i in range(0, 150):
    plt.plot(x_val, testLosses[i], marker='o')
    plt.plot(x_val, trainLosses[i], marker='x')
plt.title("Training and Test Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid(True)
plt.tight_layout()
plt.show()
```
